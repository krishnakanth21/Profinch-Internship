# Running function to find total amount transferred

```python
df2["Total amount transferred"]=""
```

```python
class Node:
    def __init__(self,sender_id,receiver_id,amount_transferred):
        self.sender = sender_id
        self.receiver = receiver_id
        self.next = None
        self.amount_transferred = amount_transferred
        self.visited = False
    def traverse(self, chain=None, total=0):
        if (self.visited): # undo visited nodes
            return
        self.visited = True
        if chain is None: # this is the beginning of the traversal
            chain = [self.sender]
        chain += [self.receiver]
        total += self.amount_transferred
        if self.next is not None:
            return self.next.traverse(chain, total)
        return chain, total
```

```python
transc = [Node( 
        row["sender_id"],
        row["receiver_id"],
        row["amount_transferred"]
    ) for i, row in df2.iterrows()]
```

```python
for i, v in enumerate(transc):
    for j, k in enumerate(transc):
        # if the receiver v same as the sender from j
        if v.receiver == k.sender:
            v.next = k


summary = [i.traverse() for i in transc]
summary = [i for i in summary if i is not None] # removing None

print(summary)
```
